```python
import os
import re

def extract_items(md_text):
    lines = md_text.split("\n")
    items = []

    current_num = None
    current_word = None
    current_def = []

    # Pattern: | 27. advocate | definition |
    row_pattern = re.compile(r"\|\s*(\d+)\.\s*([^|]+?)\s*\|\s*(.*?)\s*\|")

    for line in lines:
        match = row_pattern.search(line)

        if match:
            num, word, definition = match.groups()
            word = word.strip()
            definition = definition.replace("<br>", " ").strip()

            # Save previous entry
            if current_num is not None:
                items.append((current_num, current_word, " ".join(current_def).strip()))

            # Start new one
            current_num = num
            current_word = word
            current_def = [definition] if definition else []

        else:
            # Continuation line
            if current_num is not None:
                cont = line.replace("|", "").replace("<br>", " ").strip()
                if cont:
                    current_def.append(cont)

    # Save last
    if current_num is not None:
        items.append((current_num, current_word, " ".join(current_def).strip()))

    return items


def to_markdown(items):
    out = []
    for num, word, definition in items:
        out.append(f"### {num}. {word}\n{definition}\n")
    return "\n".join(out)


# ---------- MAIN ----------
input_filename = "manhattan_prep_1000_gre_words_.md"
output_filename = "gre_words_clean.md"

# Read from current directory
with open(input_filename, "r", encoding="utf-8") as f:
    md_text = f.read()

items = extract_items(md_text)
markdown_output = to_markdown(items)

# Write to current directory
with open(output_filename, "w", encoding="utf-8") as f:
    f.write(markdown_output)

print(f"Done! Extracted {len(items)} vocabulary items → {output_filename}")
print(f"Input read from: {os.path.abspath(input_filename)}")
print(f"Output written to: {os.path.abspath(output_filename)}")
```
